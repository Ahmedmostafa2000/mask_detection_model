```python
import keras
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os
```

```python
def predict(image):
    #plt.close()
    classes = ["without_mask","with_mask"]
    #image = cv2.imread(f"images/{image_name}")
    image = cv2.resize(image,(224, 224))
    image = np.expand_dims(image,axis = 0)
    prediction = model.predict(image)
    # printing results
    return classes[np.argmax(prediction)]
```

```python
model = keras.models.load_model("best_model.hdf5")
```

```python
files = os.listdir("images/")
f, ax = plt.subplots(len(files),1 , figsize = (45,45))
for i in range(len(files)):
    img = cv2.imread(f"images/{files[i]}")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax[i].plot()
    ax[i].axis("off")
    ax[i].set_title(files[i].split('.')[0]+"\n\n"+str(predict(img)))
    ax[i].imshow(img)
```
